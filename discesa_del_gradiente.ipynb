```python
# inizializziamo i valori
peso = 0.5
dato_input = 0.5
valore_reale = 0.8

# per quanto complicata (o semplice)
# deleghiamo la predizione a una specifica funzione
def rete_neurale(dato_input, peso):
    predizione = dato_input * peso
    return predizione


for iteration in range(134):

        predizione = rete_neurale(dato_input, peso)
        errore = (predizione - valore_reale) ** 2
        errore_puro = predizione - valore_reale
        direzione_e_ammontare = errore_puro * dato_input
        peso = peso - direzione_e_ammontare

        print(f"Iterazione: {iteration} - Errore: {errore} \t\t Predizione: {round(predizione,4)}")
```
